# Processing GDP Data

The aim of this notebook is to create 3 datasets (pandas dataframe) each containing the european country name, the GDP per capita (year 2023) and Population.

The dataframe differs for the number of countries considered, as in Level 1 of our analysis, and refer to Europe as European Union, Geographical (continental) Europe and UEFA.

The final dataframe consist of merging information extracted from the [World Bank Group](https://data.worldbank.org/indicator/NY.GDP.PCAP.CD) for the GDP and Eurostat for population. 

- `GDP_raw_data` dataframe is a time-series of 266 world county GDPs from 1960 to 2025. We will use the data from 2023 as they are not estimates but more consolided data.
- `EU_countries` dataframe consists of EU country list with their population
- `geographic_europe` consider Continental Europe
- `UEFA_countries` consider UEFA countries

```python
GDP_raw_data = pd.read_csv('./data/API_NY.GDP.PCAP.CD_DS2_en_csv_v2_46.csv',
                          skiprows=4)
print(GDP_raw_data.shape)
GDP_raw_data.head(3)
```

```python
EU_countries = pd.read_excel('./data/europe_countries_list_and_population.xlsx',
                          sheet_name='EU 27')
print(EU_countries.shape)

geographic_europe = pd.read_excel('./data/europe_countries_list_and_population.xlsx',
                          sheet_name='Geographic Europe')
print(geographic_europe.shape)

UEFA_countries = pd.read_excel('./data/europe_countries_list_and_population.xlsx',
                               sheet_name='UEFA 55')
UEFA_countries.rename(columns={'Association': 'Country'}, inplace=True)
print(UEFA_countries.shape)
```

```python
# Rename '2023' column to 'GDP' for clarity
GDP_data = GDP_raw_data[['Country Name', '2023']].rename(columns={'Country Name': 'Country', '2023': 'GDP'})

# Merge with EU_countries (preserves ALL countries in EU_countries)
EU_countries_GDP = EU_countries.merge(GDP_data, on='Country', how='left')[['Country', 'Population (M) 2023', 'GDP']]
EU_countries_GDP = EU_countries_GDP.rename(columns={'Population (M) 2023': 'Population '})

# Merge with geographic_europe (preserves ALL countries in geographic_europe)
geographic_europe_GDP = geographic_europe.merge(GDP_data, on='Country', how='left')[['Country', 'Population (M) 2023', 'GDP']]
geographic_europe_GDP = geographic_europe_GDP.rename(columns={'Population (M) 2023': 'Population '})

# Merge with UEFA_countries (preserves ALL countries in UEFA_countries)
UEFA_countries_GDP = UEFA_countries.merge(GDP_data, on='Country', how='left')[['Country', 'Population (M) 2023', 'GDP']]
UEFA_countries_GDP = UEFA_countries_GDP.rename(columns={'Population (M) 2023': 'Population '})

print("EU_countries_GDP shape:", EU_countries_GDP.shape)
print("geographic_europe_GDP shape:", geographic_europe_GDP.shape)
print("UEFA_countries_GDP shape:", UEFA_countries_GDP.shape)
print("\nEU_countries_GDP:")
print(EU_countries_GDP.head())
print("\nMissing GDP values in EU_countries_GDP:", EU_countries_GDP['GDP'].isna().sum())
print("\ngeographic_europe_GDP:")
print(geographic_europe_GDP.head())
print("\nMissing GDP values in geographic_europe_GDP:", geographic_europe_GDP['GDP'].isna().sum())
print("\nUEFA_countries_GDP:")
print(UEFA_countries_GDP.head())
print("\nMissing GDP values in UEFA_countries_GDP:", UEFA_countries_GDP['GDP'].isna().sum())
```

In `GDP_raw_data` there is a mismatch between how 'Czech Republic' and 'Slovakia' are named in the other dataframe. Let's fix this by taking the '2023' values for both `Czechia` and `Slovak Republic` in `GDP_raw_data` and fill the other dataframe namely `EU_countries_GDP`, `geographic_europe_GDP`, `UEFA_countries_GDP`

- Strategy to Fix the Naming Mismatch

We'll extract the GDP values for 'Czechia' and 'Slovak Republic' from `GDP_raw_data`, then use these values to fill the missing GDP data in our three dataframes (`EU_countries_GDP`, `geographic_europe_GDP`, and `UEFA_countries_GDP`). This approach ensures we have consistent country naming across all datasets while preserving the economic data.

```python
# Extract GDP values for 'Czechia' and 'Slovak Republic' from GDP_raw_data
czechia_gdp = GDP_raw_data[GDP_raw_data['Country Name'] == 'Czechia']['2023'].values[0]
slovak_republic_gdp = GDP_raw_data[GDP_raw_data['Country Name'] == 'Slovak Republic']['2023'].values[0]

print(f"Czechia GDP (2023): {czechia_gdp}")
print(f"Slovak Republic GDP (2023): {slovak_republic_gdp}")

# Fill missing GDP values in EU_countries_GDP
EU_countries_GDP.loc[EU_countries_GDP['Country'] == 'Czech Republic', 'GDP'] = czechia_gdp
EU_countries_GDP.loc[EU_countries_GDP['Country'] == 'Slovakia', 'GDP'] = slovak_republic_gdp

# Fill missing GDP values in geographic_europe_GDP
geographic_europe_GDP.loc[geographic_europe_GDP['Country'] == 'Czech Republic', 'GDP'] = czechia_gdp
geographic_europe_GDP.loc[geographic_europe_GDP['Country'] == 'Slovakia', 'GDP'] = slovak_republic_gdp

# Fill missing GDP values in UEFA_countries_GDP
UEFA_countries_GDP.loc[UEFA_countries_GDP['Country'] == 'Czech Republic', 'GDP'] = czechia_gdp
UEFA_countries_GDP.loc[UEFA_countries_GDP['Country'] == 'Slovakia', 'GDP'] = slovak_republic_gdp

print("\nAfter filling name mismatches:")
print(f"Missing GDP values in EU_countries_GDP: {EU_countries_GDP['GDP'].isna().sum()}")
print(f"Missing GDP values in geographic_europe_GDP: {geographic_europe_GDP['GDP'].isna().sum()}")
print(f"Missing GDP values in UEFA_countries_GDP: {UEFA_countries_GDP['GDP'].isna().sum()}")

print("\nUpdated EUR_countries_GDP:")
print(EU_countries_GDP[EU_countries_GDP['Country'].isin(['Czech Republic', 'Slovakia'])])
print("\nUpdated geographic_europe_GDP:")
print(geographic_europe_GDP[geographic_europe_GDP['Country'].isin(['Czech Republic', 'Slovakia'])])
print("\nUpdated UEFA_countries_GDP:")
print(UEFA_countries_GDP[UEFA_countries_GDP['Country'].isin(['Czech Republic', 'Slovakia'])])
```

As for `geographic_europe_GDP` I have a similar issue where in `geographic_europe_GDP` the following countries 'Bosnia & Herzegovina', 'Russia', 'Turkey' are named in `GDP_raw_data` as `Bosnia and Herzegovina`, `Russian Federation` and `Turkiye`

```python
# Extract GDP values for the mismatched country names from GDP_raw_data
bosnia_gdp = GDP_raw_data[GDP_raw_data['Country Name'] == 'Bosnia and Herzegovina']['2023'].values[0]
russia_gdp = GDP_raw_data[GDP_raw_data['Country Name'] == 'Russian Federation']['2023'].values[0]
turkey_gdp = GDP_raw_data[GDP_raw_data['Country Name'] == 'Turkiye']['2023'].values[0]

print(f"Bosnia and Herzegovina GDP (2023): {bosnia_gdp}")
print(f"Russian Federation GDP (2023): {russia_gdp}")
print(f"Turkiye GDP (2023): {turkey_gdp}")

# Fill missing GDP values in geographic_europe_GDP
geographic_europe_GDP.loc[geographic_europe_GDP['Country'] == 'Bosnia & Herzegovina', 'GDP'] = bosnia_gdp
geographic_europe_GDP.loc[geographic_europe_GDP['Country'] == 'Russia', 'GDP'] = russia_gdp
geographic_europe_GDP.loc[geographic_europe_GDP['Country'] == 'Turkey', 'GDP'] = turkey_gdp

# Fill missing GDP values in UEFA_countries_GDP
UEFA_countries_GDP.loc[UEFA_countries_GDP['Country'] == 'Bosnia & Herzegovina', 'GDP'] = bosnia_gdp
UEFA_countries_GDP.loc[UEFA_countries_GDP['Country'] == 'Russia', 'GDP'] = russia_gdp
UEFA_countries_GDP.loc[UEFA_countries_GDP['Country'] == 'Turkey', 'GDP'] = turkey_gdp

print("\nAfter filling geographic_europe naming mismatches:")
print(f"Missing GDP values in geographic_europe_GDP: {geographic_europe_GDP['GDP'].isna().sum()}")
print(f"Missing GDP values in UEFA_countries_GDP: {UEFA_countries_GDP['GDP'].isna().sum()}")

print("\nUpdated geographic_europe_GDP:")
print(geographic_europe_GDP[geographic_europe_GDP['Country'].isin(['Bosnia & Herzegovina', 'Russia', 'Turkey'])])
print("\nUpdated UEFA_countries_GDP:")
print(UEFA_countries_GDP[UEFA_countries_GDP['Country'].isin(['Bosnia & Herzegovina', 'Russia', 'Turkey'])])
```

We are almost there. As for Gibraltar, we can use the [latest data](https://www.gibraltar.gov.gi/statistics/key-indicators) and use a value £80885 and convert into 2023 US$. 

A representative reference for 2023 average FX rates is 2023 average GBP to USD ≈ 1.2440 USD per £1 (average for the year). This is a reasonable benchmark for converting non-USD amounts to USD for comparisons during that period. Using this values, I will set a 100,603 value for the GDP in the dataframe

```python
# Fill Gibraltar with the specified value
gibraltar_gdp = 100603

# Extract UK GDP value for England, Northern Ireland, Scotland, and Wales
uk_gdp = GDP_raw_data[GDP_raw_data['Country Name'] == 'United Kingdom']['2023'].values[0]

print(f"Gibraltar GDP (2023): {gibraltar_gdp}")
print(f"United Kingdom GDP (2023): {uk_gdp}")

# Fill Gibraltar in geographic_europe_GDP
geographic_europe_GDP.loc[geographic_europe_GDP['Country'] == 'Gibraltar', 'GDP'] = gibraltar_gdp

# Fill Gibraltar in UEFA_countries_GDP
UEFA_countries_GDP.loc[UEFA_countries_GDP['Country'] == 'Gibraltar', 'GDP'] = gibraltar_gdp

# Fill UK constituent countries (England, Northern Ireland, Scotland, Wales) in UEFA_countries_GDP
uk_countries = ['England', 'Northern Ireland', 'Scotland', 'Wales']
for country in uk_countries:
    UEFA_countries_GDP.loc[UEFA_countries_GDP['Country'] == country, 'GDP'] = uk_gdp

print("\nAfter filling Gibraltar and UK constituent countries:")
print(f"Missing GDP values in geographic_europe_GDP: {geographic_europe_GDP['GDP'].isna().sum()}")
print(f"Missing GDP values in UEFA_countries_GDP: {UEFA_countries_GDP['GDP'].isna().sum()}")

print("\nUpdated geographic_europe_GDP (Gibraltar):")
print(geographic_europe_GDP[geographic_europe_GDP['Country'] == 'Gibraltar'])

print("\nUpdated UEFA_countries_GDP (Gibraltar and UK countries):")
print(UEFA_countries_GDP[UEFA_countries_GDP['Country'].isin(['Gibraltar', 'England', 'Northern Ireland', 'Scotland', 'Wales'])])
```
